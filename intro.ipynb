```python
import pandas as pd

# Load the original Excel file again
df = pd.read_excel('SuperStore Sales DataSet.xlsx')

# Show shape and top 5 rows
print("Data shape:", df.shape)
df.head()
```

```python
# Clean column names
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(" ", "_")
              .str.replace("-", "_")
)

# Drop empty or irrelevant columns
df.drop(columns=['row_id+o6g3a1:r6', 'returns'], inplace=True, errors='ignore')

# Convert dates
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')

# Show cleaned data
df.head()
```

```python
total_sales = df['sales'].sum()
total_profit = df['profit'].sum()
total_quantity = df['quantity'].sum()

print("💰 Total Sales: ₹", round(total_sales, 2))
print("📦 Total Quantity Sold:", total_quantity)
print("💸 Total Profit: ₹", round(total_profit, 2))
```

```python
# Grouping total sales by product category
category_sales = df.groupby('category')['sales'].sum().sort_values(ascending=False)

# Displaying the result
print(category_sales)
Data
```

```python
import matplotlib.pyplot as plt

# Top 10 Products by Sales
top_products_sales = df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)
print("🔝 Top 10 Products by Sales:\n", top_products_sales)

# Plot bar chart for top products by sales
top_products_sales.plot(kind='barh', color='lightgreen', title='Top 10 Products by Sales')
plt.xlabel('Sales (₹)')
plt.ylabel('Product')
plt.gca().invert_yaxis()  # Highest sales at the top
plt.tight_layout()
plt.show()

# Top 10 Products by Profit
top_products_profit = df.groupby('product_name')['profit'].sum().sort_values(ascending=False).head(10)
print("💸 Top 10 Products by Profit:\n", top_products_profit)

# Plot bar chart for top products by profit
top_products_profit.plot(kind='barh', color='lightblue', title='Top 10 Products by Profit')
plt.xlabel('Profit (₹)')
plt.ylabel('Product')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Group sales by region
sales_by_region = df.groupby('region')['sales'].sum().sort_values(ascending=False)
print("📍 Total Sales by Region:\n", sales_by_region)

# Plot bar chart
sales_by_region.plot(kind='bar', color='orange', title='Total Sales by Region')
plt.ylabel('Sales (₹)')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```

```python
# Group sales by state
sales_by_state = df.groupby('state')['sales'].sum().sort_values(ascending=False)
print("📍 Total Sales by State:\n", sales_by_state)

# Plot bar chart (top 10 states by sales)
sales_by_state.head(10).plot(kind='bar', color='purple', title='Top 10 States by Sales')
plt.ylabel('Sales (₹)')
plt.xlabel('State')
plt.tight_layout()
plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['year_month'] = df['order_date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('year_month')['sales'].sum()
print("📅 Monthly Sales Trend:\n", monthly_sales)
```

```python
monthly_profit = df.groupby('year_month')['profit'].sum()
monthly_profit.plot(kind='line', marker='o', color='green', title='Monthly Profit Trend')
plt.ylabel('Profit (₹)')
plt.xlabel('Year-Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['quarter'] = df['order_date'].dt.to_period('Q')
df['month'] = df['order_date'].dt.month

```

```python
sales_by_quarter = df.groupby('quarter')['sales'].sum()
print("📊 Sales by Quarter:\n", sales_by_quarter)

# Plot quarterly sales
sales_by_quarter.plot(kind='bar', color='coral', title='Sales by Quarter')
plt.ylabel('Sales (₹)')
plt.xlabel('Quarter')
plt.tight_layout()
plt.show()
```

```python
sales_by_month = df.groupby('month')['sales'].sum()
print("📊 Sales by Month (All Years Combined):\n", sales_by_month)

# Plot monthly sales to detect seasonality
sales_by_month.plot(kind='bar', color='skyblue', title='Sales by Month (All Years Combined)')
plt.ylabel('Sales (₹)')
plt.xlabel('Month')
plt.xticks(ticks=range(1,13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()
```
